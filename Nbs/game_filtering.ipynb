```python
import pandas as pd
```

```python
fp = "popular_board_game.csv"
df = pd.read_csv(fp)
```

```python
def reduction(row):
    if row==1:
        return 'solo'
    return 'team'
df['team'] = df['min_players'].apply(reduction)
```
